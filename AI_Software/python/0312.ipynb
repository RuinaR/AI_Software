```python
#string
str = "ABCDEFG"
print(str)
str = '''multiline
string'''
print(str)


print(str[3])
print(str[3:])

#enumerate() <- 리터럴의 모음을 튜플 덩어리로 만듬(list, string 등)
list_enumerate = list(enumerate(str)) #튜플 덩어리를 리스트로 변환
print("list_enumerate = ", list_enumerate)
```
