```python
#namespace
#global : 전역 공간의 같은이름의 전역변수를 가져온다는 의미

def outerFunc():
    global a
    a = 20
    print(a)
    def innerFunc():
        global a
        a = 30
        print('a=', a)
    innerFunc()
    print('a=', a)

#전역 변수 공간
a = 10
print(a)
outerFunc()
print('a=', a)
```
